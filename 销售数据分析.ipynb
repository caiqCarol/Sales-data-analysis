```python
#把str字符串全部改成小写字母
df.columns = df.columns.str.lower()
#把列名中的空格''改成下横线'-'
df.columns = df.columns.str.replace(' ','_')
#查看
df.head(2)
```

```python
#增加新列discount，并计算折扣金额
df['discount'] = df['list_price'] * df['discount_percent'] * .01
#增加新列sale_price,并计算销售价格
df['sale_price'] = df['list_price'] - df['discount']
#增加新列profit，并计算利润
df['profit'] = df['sale_price'] - df['cost_price']
#查看
df.head(5)
```

```python
#order_date这列从字符串转换成日期格式年月日    目的:把字符串转换成真正可计算的日期格式
df['order_date'] = pd.to_datetime(df['order_date'], format='%Y-%m-%d')
#查看
df.dtypes

#上面转换日期格式后可以
#1提取年份
# df['order_date'].dt.year
#2提取月份
# df['order_date'].dt.month
#3提取时间差 如果表格里除了预定时间还有购物时间的话，可以计算时间差。 本表格没有购物时间
# df['order_date'] - df['ship_date']
#4按时间排序
# df.sort_values('order_date')
```

```python
#删除这三列，inplace=true在原表格上删除列
df.drop(columns = ['cost_price', 'list_price', 'discount_percent'], inplace=True)
df.head(2)
```

```python
#用 SQLAlchemy 连接 SQL Server 数据库

#数据库类型：mssql服务器：Josh数据库名：Projects驱动：SQL Server它创建了一个“连接引擎”

#真正建立连接
#连接数据库
import sqlalchemy as sa
#数据库在本机所以写.\SQLEXPRESS
SERVER_NAME = r".\SQLEXPRESS"
#创建了一个叫做Projects的数据库
DB_NAME = "Projects"
#创建一个数据库连接口
engine = sa.create_engine(
    f"mssql+pyodbc://@{SERVER_NAME}/{DB_NAME}"   #数据库类型 + 驱动  ://  用户@服务器/数据库  ?参数
    "?driver=ODBC+Driver+17+for+SQL+Server"
    "&trusted_connection=yes"
    "&TrustServerCertificate=yes"
)

with engine.connect() as conn:
    print(conn.execute(sa.text("SELECT 1")).fetchall())
```

```python
#把 pandas 的 df 写进数据库   把 pandas 的 DataFrame写入 SQL Server,表名叫 df_orders
#to_sql把数据写进数据库的一张表里   往数据库表里写入 38 条记录 有 38 行数据被成功写入 SQL Server 里的 df_orders 表

df.to_sql('df_orders', con=engine, index=False, if_exists='replace')
```
